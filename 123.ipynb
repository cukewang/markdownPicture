```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('hebust(1).xls')
df.head()
```

```python
df1 = df.groupby('tj_yxsmc').agg([np.sum,np.max,np.min])['总成绩（初试+复试）'].reset_index()
df1
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
df1.plot.bar(x='tj_yxsmc',y=['amax','amin'],
            title=('不同系别最高最低分对比'),
            figsize=(10,6)
            )
# plt.bar(df1.tj_yxsmc,df1['amax'])
# plt.bar(df1.tj_yxsmc,df1['amin'])
plt.xticks(rotation=45)
```
